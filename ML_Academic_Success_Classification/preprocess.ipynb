```python
def preprocess_data(data, is_train=True):
    # pop id
    ids = data.pop('id')

    drop_columns = [
        'Nacionality',
        'International']
    
    data = data.drop(columns=drop_columns)
    
    # Create interaction features
    data['Interaction_Curricular_units'] = data['Curricular units 1st sem (enrolled)'] * data['Curricular units 2nd sem (enrolled)']
    
    # Create ratio features
    data['Ratio_Approved_Enrolled_1st'] = data['Curricular units 1st sem (approved)'] / (data['Curricular units 1st sem (enrolled)'] + 1e-5)
    data['Ratio_Approved_Enrolled_2nd'] = data['Curricular units 2nd sem (approved)'] / (data['Curricular units 2nd sem (enrolled)'] + 1e-5)
    
    # # Create new feature: Cumulative Grades
    data['Cumulative_Grades'] = data['Curricular units 1st sem (grade)'] + data['Curricular units 2nd sem (grade)']
    
    # # Handle missing values for numeric columns
    # numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
    # data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].median())
    
    # # Convert categorical variables into dummy/indicator variables
    # categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
    # categorical_cols = [col for col in categorical_cols if col != 'Target']  # Exclude the target column from dummy conversion
    # data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
    
    if is_train:
        target = data.pop('Target')  # pop the target variable for training data
        return data, target, ids
    else:
        return data, ids
```

```python
# Preprocess the datasets
df_train, y_train, train_ids = preprocess_data(train_data, is_train=True)
df_test, test_ids = preprocess_data(test_data, is_train=False)
```
